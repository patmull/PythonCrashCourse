```python
figure, axis = plt.subplots()

x_values = [1,2,3,4,5]
y_values = [x_val**2 for x_val in x_values]
del y_values[0]

same_length = len(x_values) == len(y_values)
print(f"Length same? {same_length}")

try:
    axis.plot(x_values, y_values, linewidth=4)
    plt.show()
except ValueError as ve:
    if "x and y must have same first dimension" in str(ve):
        print("Invalid dimensions!")
```
